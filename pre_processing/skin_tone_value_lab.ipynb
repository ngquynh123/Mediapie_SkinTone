```python
import os
import cv2
import numpy as np
import pandas as pd

def get_average_lab(image_bgr):
    lab = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2Lab)
    L, a, b = cv2.split(lab)
    return np.mean(L), np.mean(a), np.mean(b)

def build_skin_tone_template_from_images(sample_dir):
    template = {}
    valid_exts = ('.jpg', '.jpeg', '.png', '.bmp')

    for type_name in sorted(os.listdir(sample_dir)):
        type_path = os.path.join(sample_dir, type_name)
        if not os.path.isdir(type_path):
            continue

        L_list, a_list, b_list = [], [], []
        found_image = False

        print(f"\nğŸ” Äang xá»­ lÃ½: {type_name}")

        for fname in os.listdir(type_path):
            if fname.lower().endswith(valid_exts):
                fpath = os.path.join(type_path, fname)
                img = cv2.imread(fpath)
                if img is not None:
                    found_image = True
                    L, a_, b_ = get_average_lab(img)
                    L_list.append(L)
                    a_list.append(a_)
                    b_list.append(b_)
                    print(f"    ÄÃ£ Ä‘á»c: {fname}")
                else:
                    print(f"    KhÃ´ng Ä‘á»c Ä‘Æ°á»£c áº£nh: {fname}")
        
        if found_image:
            L_avg = np.mean(L_list)
            a_avg = np.mean(a_list)
            b_avg = np.mean(b_list)
            template[type_name] = [round(L_avg, 2), round(a_avg, 2), round(b_avg, 2)]
            print(f"   â¤ Trung bÃ¬nh: L={L_avg:.2f}, a={a_avg:.2f}, b={b_avg:.2f}")
        else:
            print(f"    KhÃ´ng tÃ¬m tháº¥y áº£nh há»£p lá»‡ trong {type_path}")

    return template
```

```python
sample_dir = "D:/KLTN/SKINTONE/public/skin tone values"  # ÄÆ°á»ng dáº«n thÆ° má»¥c báº¡n Ä‘Ã£ gá»­i

skin_tone_templates = build_skin_tone_template_from_images(sample_dir)

# LÆ°u ra file Ä‘á»ƒ kiá»ƒm tra báº£ng máº«u
df_template = pd.DataFrame([
    {"type": k, "L": v[0], "a": v[1], "b": v[2]} for k, v in skin_tone_templates.items()
])
df_template.to_csv("skin_tone_template_from_samples.csv", index=False)
print("\n ÄÃ£ lÆ°u báº£ng máº«u tone da tá»« áº£nh vÃ o: skin_tone_template_from_samples.csv")
```
